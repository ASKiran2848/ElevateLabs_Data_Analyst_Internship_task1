```python
import kagglehub
import pandas as pd
import numpy as np
import os
```

```python
path = kagglehub.dataset_download("kyanyoga/sample-sales-data")
print("Dataset downloaded to:", path)
```

```python
print("Files inside dataset folder:", os.listdir(path))
```

```python
csv_file = "sales_data_sample.csv"
```

```python
df = pd.read_csv(path + "/" + csv_file, encoding="latin1")
print("\nPreview of dataset:")
print(df.head())
```

```python
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")
```

```python
print("\nMissing values:\n", df.isnull().sum())
print("\nDuplicate rows:", df.duplicated().sum())
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df = df.drop_duplicates()
```

```python
if "orderdate" in df.columns:
    df["orderdate"] = pd.to_datetime(df["orderdate"], errors="coerce")
```

```python
expected_numeric = ["quantityordered", "priceeach", "sales"]
for col in expected_numeric:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce").fillna(0)
```

```python
text_cols = df.select_dtypes(include='object').columns
for col in text_cols:
    df[col] = df[col].str.strip().str.title()
```

```python
if "sales" in df.columns:
    df = df[df["sales"] < df["sales"].quantile(0.99)]
```

```python
import os

save_path = "/content/sales_output"   # for Google Colab
os.makedirs(save_path, exist_ok=True)

df.to_csv(save_path + "/sales_data_cleaned.csv", index=False)
print("Saved to:", save_path)
```
